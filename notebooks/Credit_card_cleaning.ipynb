```python
customer_metrics = transactions.groupby('client_id').agg({'amount':['sum','mean','count','std']})
```

```python
customer_metrics.columns = [
    'total_spend',
    'avg_spend',
    'transaction_count',
    'spending_volatility'
]
```

```python
customer_metrics = customer_metrics.reset_index()
```

```python
customer_metrics.head()
```

```python
customer_metrics.describe()
```

```python
customer_metrics['engagement_score'] = (customer_metrics['transaction_count'] * 0.6 + customer_metrics['avg_spend'] * 0.4)
```

```python
customer_metrics['spending_tier'] = pd.cut(
    customer_metrics['total_spend'],
    bins=[0,3000,7000,15000,40000],
    labels=[
        'Low Spender',
        'Moderate Spender',
        'High Spender',
        'Premium Spender'
    ]
)
```

```python
customer_metrics['spending_tier'].value_counts()
```

```python
customer_metrics['volatility_level'] = pd.cut(
    customer_metrics['spending_volatility'],
    bins=[0,50,200,500,100000],
    labels=[
        'Stable',
        'Moderate',
        'Irregular',
        'Highly Volatile'
    ]
)
```

```python
customer_metrics['volatility_level'].value_counts()
```

```python
customer_metrics['wellness_score'] = (customer_metrics['avg_spend'] * 0.3 + customer_metrics['transaction_count'] * 0.4 - customer_metrics['spending_volatility'] * 0.2)
```

```python
customer_metrics[
    [
        'client_id',
        'wellness_score',
        'transaction_count',
        'avg_spend',
        'spending_volatility'
    ]
].head()
```

```python
customer_metrics['wellness_category'] = pd.cut(
    customer_metrics['wellness_score'],
    bins=[-1000,50,100,200,10000],
    labels=[
        'Low Wellness',
        'Moderate Wellness',
        'Healthy',
        'Excellent'
    ]
)
```

```python
customer_metrics['wellness_category'].value_counts()
```

```python
customer_metrics['high_value_customer'] = np.where(customer_metrics['total_spend'] > 15000,1,0)
```

```python
digital_usage = transactions.groupby('client_id')['use_chip'].apply(lambda x: (x.str.contains('Online', case=False)).mean()).reset_index()
```

```python
digital_usage.columns = [
    'client_id',
    'digital_transaction_ratio'
]
```

```python
customer_metrics = customer_metrics.merge(
    digital_usage,
    on='client_id',
    how='left'
)
```

```python
customer_metrics['digital_persona'] = np.where(
    customer_metrics['digital_transaction_ratio'] > 0.5,
    'Digital First',
    'Traditional User'
)
```

```python
customer_metrics['digital_persona'].value_counts()
```

```python
last_transaction = transactions.groupby('client_id')['date'].max().reset_index()
```

```python
last_transaction.columns = [
    'client_id',
    'last_transaction_date'
]
```

```python
customer_metrics = customer_metrics.merge(
    last_transaction,
    on='client_id',
    how='left'
)
```

```python
latest_date = transactions['date'].max()
customer_metrics['days_since_last_transaction'] = (latest_date - customer_metrics['last_transaction_date']).dt.days
```

```python
customer_metrics['churn_risk'] = np.where(
    customer_metrics['days_since_last_transaction'] > 30,
    'High Risk',
    'Active'
)
```

```python
customer_metrics['churn_risk'].value_counts()
```

```python
customer_analytics = customer_metrics.merge(
    users,
    left_on='client_id',
    right_on='id',
    how='left'
)
```
